```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
'Age': [25, 30, 35, 40],
'Score': [85, 90, 75, 883]}
df = pd.DataFrame(data)
```

```python
df
```

```python
#Boolean array for filtering
sample_condition = df['Age']>30
df.loc[sample_condition]
```

```python
multiple_condition = (df[ 'Age'] >= 30) & (df[ 'Score'] > 80)
```

```python
df.loc[multiple_condition]
```

```python
df.loc[df[ 'Age' ]>=30, 'Score']= 100
```

```python
df
```

```python
condition = df[ 'Score'] == 100
df. loc[ condition, 'Name']
```

```python
df.loc[condition.['Name','Age']]
```
